Ø Create a histogram for the feature ‘KM’.


·        Also set the following properties:

o   No of bins=5

o   Edge color=White




o   X label= Kilometer

o   Y label= Frequency


o   Title= Histogram of Kilometer

```python
plt.hist(x=toyota['KM'],color='purple',bins=5,edgecolor='white')
plt.title("Histogram of Kilometer")
plt.xlabel("saf")
plt.ylabel("yerw")
plt.show()
```

```python
histogram = sns.displot(toyota['KM'],bins=5,edgecolor='white')
histogram.set_axis_labels('KM', 'Frequency')
```
