```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
# combinig both df and df-country dataset
final_df = pd.merge(df,df_country,on='Country Code', how='left')
final_df.head()
```

```python
# to check datatypes
final_df.dtypes
```

```python
final_df.columns
```

```python
country_names = final_df.Country.value_counts().index
```

```python
country_values = final_df.Country.value_counts().values
```

```python
# pie chart

plt.pie(country_values,labels = country_names)
plt.show()
```

```python
# pie chart - Top 3 countries that uses zomato

plt.pie(country_values[:3],labels = country_names[:3], autopct='%1.2f%%')
plt.show()
```

Observation:

    Zomato maximum records or transaction are from India.    
    After that USA and then UK
        

```python
final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
```

```python
ratings = final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
```

```python
ratings
```

### Observation
1. When ratings are between 4.5 to 4.9  --> Excellent
2. When ratings are between 4.0 to 4.4  --> Very good
3. When ratings are between 3.5 to 3.9  --> good
4. When ratings are between 2.5 to 3.4  --> average
5. When ratings are between 2.0 to 2.4  --> poor

```python
ratings.head()
```

```python
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(x = 'Aggregate rating', y = 'Rating count',data=ratings)
```

```python
sns.barplot(x = 'Aggregate rating', y = 'Rating count',hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','darkgreen'])
```

Observations:
    
        1. Not rated count is very high
        2. Maximum number of ratings are between 2.5 to 3.4

```python
# count plot
sns.countplot(x='Rating color',data=ratings, palette=['blue','red','orange','yellow','green','darkgreen'])
```
